```python
import numpy as np

file_path = 'notebooks/Phase 4/phase_4_embeddings/F1_P1/test/299811.npy'
try:
    # Load the data from the .npy file
    data = np.load(file_path)
    
    # Print information about the loaded data
    print("Successfully loaded the NumPy file.")
    print("Shape of the array:", data.shape)
    print("Data type of the array:", data.dtype)
    
    # Print the first 10 elements as an example
    print("\nFirst 10 elements of the array:")
    print(data[:10])
    
except FileNotFoundError:
    print(f"Error: The file was not found at the path: {file_path}")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
```
